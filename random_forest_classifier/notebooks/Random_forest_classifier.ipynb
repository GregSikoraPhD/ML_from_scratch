```python
from random_forrest_cls import accuracy, RandomForrestCls
```

```python
data = datasets.load_iris()
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
clf = RandomForrestCls(n_trees=50)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict_value(X_test)
acc = accuracy(y_test, y_pred)
print(acc)
```
